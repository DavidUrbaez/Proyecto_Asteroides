SNR calculation

```python
r=10
x=120-35
y=300+10+5
zoom_img=datos[x-r:x+r+1,y-r:y+r+1]
plt.imshow(zoom_img,cmap='gray',vmax=2000)

circle1 = plt.Circle((r, r), r/2, color='r', fill=False, lw=4) # img[25,55] <- center
plt.gca().add_patch(circle1)
```

```python
import numpy as np
distances=[]
value=[]
for i in range(zoom_img.shape[0]):
  for j in range(zoom_img.shape[1]):
    dis=np.sqrt((i-10)**2+(j-10)**2)
    if dis<10:
      distances.append(dis)
      value.append(zoom_img[i,j])
```

```python
plt.style.available
plt.style.use('dark_background')
plt.scatter(distances,value,c=value,cmap='winter')
plt.xlabel('Distacia al centro [px]')
plt.ylabel('Intensidad de la imagen')
```
